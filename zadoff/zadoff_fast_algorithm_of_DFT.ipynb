$$
x_u(m) = e^{-j\frac{\pi u m (m+1) }{L}} \tag{1.1}
$$

这是一个长度为 L 的 zadoff 序列，起根为 u，其中 m 的取值范围为 [0,1,...,L-1].

```python
import matplotlib.pyplot as plt
import numpy as np

L = 7
u = 3

i = np.arange(0,L,1)
zadoff = np.exp(-1j*np.pi*u*i*(i+1)/L)
fig,[ax_re,ax_img]=plt.subplots(1,2)
ax_re.plot(np.real(zadoff),".--")
ax_img.plot(np.imag(zadoff),".--")
fig.show()
```

如果对上面的 $x_u(m)$ 序列做 DFT 变换，公式如下：

$$
y(k)=\sum_{m=0}^{L-1} x_u(m) e^{-j 2\pi \frac{k}{L}m}  \tag{1.2}
$$

# zadoff 公式的周期性
性质:
**公式 (1.1) 是以 L 为周期的**
